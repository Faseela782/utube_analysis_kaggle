```python
#ipywidgets
```

```python
import ipywidgets as widgets
```

```python
c=widgets.Dropdown(
    options=video['genre'].unique(),
    value='Music',
    description='genre',
    disabled=False,
)
def topchannel(a):
  channel=video[video['genre']== a]
  final=channel.select_dtypes(include=['number', 'object']).groupby(by='channelTitle').sum().sort_values(by='view_count',ascending=False).head(5)
  sns.barplot(y=final.index,x=final['view_count'], estimator=sum)

  plt.title(a)
  plt.show()

widgets.interact(topchannel,a=c)
```
